```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f1(x, y):
    c = 1.19
    d = 0.63
    return y*(d*x-c)
def f2(x, y):
    a = 1.58
    b = 1.94
    return x*(a-b*y)
```

```python
epsilon = 10**(-3)
x0 = 2.85
y0 = 1.85
t0 = 9
tn = 29
dt = tn - t0
n = int((dt/(epsilon))**(1/4) + 1)
k = 1
i = 1
while True:
    y_ = [y0]
    x_ = [x0]
    nk = n*2**k
    h = round(dt/(nk), 5)
    print("кількість розбитів: n = {}\nкрок: h = {}".format(nk, h))
    k1 = h*f1(x_[i-1], y_[i-1])
    q1 = h*f2(x_[i-1], y_[i-1])
    k2 = h*f1(x_[i-1]+q1/2, y_[i-1]+k1/2)
    q2 = h*f2(x_[i-1]+q1/2, y_[i-1]+k1/2)
    k3 = h*f1(x_[i-1]+q2/2, y_[i-1]+k2/2)
    q3 = h*f2(x_[i-1]+q2/2, y_[i-1]+k2/2)
    k4 = h*f1(x_[i-1]+q3, y_[i-1]+k3)
    q4 = h*f2(x_[i-1]+q3, y_[i-1]+k3)
    y_.append(y_[i-1]+(k1+2*k2+2*k3+k4)/6)
    x_.append(x_[i-1]+(q1+2*q2+2*q3+q4)/6)
    if abs(x_[1] - x_[0])/15 <= epsilon and abs(y_[1] - y_[0])/15 <= epsilon:
        break
    else: k += 1
print("\n\n")
print("кількість розбитів що задовільняє потрібну точність: n = {}\nкрок що задовільняє потрібну точність: h = {}".format(nk, h))
```

```python
y = [y0]
x = [x0]
for i in range(1, nk+1):
    k1 = h*f1(x[i-1], y[i-1])
    q1 = h*f2(x[i-1], y[i-1])
    k2 = h*f1(x[i-1]+q1/2, y[i-1]+k1/2)
    q2 = h*f2(x[i-1]+q1/2, y[i-1]+k1/2)
    k3 = h*f1(x[i-1]+q2/2, y[i-1]+k2/2)
    q3 = h*f2(x[i-1]+q2/2, y[i-1]+k2/2)
    k4 = h*f1(x[i-1]+q3, y[i-1]+k3)
    q4 = h*f2(x[i-1]+q3, y[i-1]+k3)
    y.append(y[i-1]+(k1+2*k2+2*k3+k4)/6)
    x.append(x[i-1]+(q1+2*q2+2*q3+q4)/6)
```

```python
plt.figure(figsize = (16, 10))
plt.grid()
plt.plot(np.linspace(t0, tn, nk+1), y, label = "Популяція Жертв")
plt.plot(np.linspace(t0, tn, nk+1), x, label = "Популяція Хижаків")
plt.xlabel('t')
plt.ylabel('y, x')
plt.title('Популяція')
plt.legend(loc=2)
```
