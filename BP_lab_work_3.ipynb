```python
#task 5
def get_state(pn, p):
    F = pn[0]
    i = 0
    while(F < p):
        i = i + 1
        F = F + pn[i]
    return i
def eye(dimens):
    return np.diag(np.diag([[1]*dimens]*dimens))
def get_V(P, m):
    P_size = len(P[0, :])
    V = np.array([[0]*P_size]*P_size)
    P_pow = eye(P_size)
    for i in range(m):
        P_pow = np.dot(P_pow, P)
        V = V + P_pow
    return V
def get_M (Z, f):
    size = len(f)
    I = eye(size)
    E = np.array([[1]*size]*size)
    DZ = np.diag(np.diag(Z))
    DM = np.diag(np.diag([[1/size]*size]*size))
    M = np.dot((I - Z + np.dot(E, DZ)), DM)
    return M
  
n = 5
p01 = [0, 0, 1, 0, 0]
p02 = [0.1, 0.2, 0.3, 0.3, 0.1]
m = 100
states1 = []
states2 = []
P = np.array([[0.2,0.1,0.1,0.4,0.2],[0.2,0.1,0.5,0,0.2],[0.6,0.1,0.1,0.1,0.1],[0.4,0,0.5,0,0.1],[0.4,0.3,0,0.3,0]])
#P = np.array([[1/n]*n]*n)
ps1 = np.random.uniform(0, 1, m + 1)
ps2 = np.random.uniform(0, 1, m + 1)
pn1 = []
pn2 = []
P_size = len(P[0, :])
F = P
for i in range(100):
    F = np.dot(F, P)
f = F[0, :]
Z = np.linalg.inv(eye(P_size) - (P - F))
V = get_V(P, m)
M = get_M(Z, f)
tn1 = np.dot(p01,V)
t1 = np.dot(p01,M)
states1.append(get_state(p01, ps1[0]))
tn_exp1 = [i for i in range(P_size )]
t_exp1 = [i for i in range(P_size )]
for i in range(m):
    states1.append(get_state(P[states1[i], :], ps1[i + 1]))
    tn_exp1[states1[i]] = tn_exp1[states1[i]] + 1
    if(t_exp1[states1[i]] == 0):
        t_exp1[states1[i]] = i
tn2 = np.dot(p02,V)
t2 = np.dot(p02,M)
states2.append(get_state(p02, ps2[0]))
tn_exp2 = [i for i in range(P_size )]
t_exp2 = [i for i in range(P_size )]
for i in range(m):
    states2.append(get_state(P[states2[i], :], ps2[i + 1]))
    tn_exp2[states2[i]] = tn_exp2[states2[i]] + 1
    if(t_exp2[states2[i]] == 0):
        t_exp2[states2[i]] = i
print("Фінальна матриця F:")
print(F)
print("Z:")
print(Z)
print("Середня тривалість переміщення між відповідними станами:")
print(V)
print("Середні моменти першого трансферу між відповідними станами:")
print(M)
print("Точний стан:")
print("Середня тривалість перебування в відповідних станах:")
print(tn1)
print("Середні моменти першого прибуття в відповідні стани:")
print(t1)
print("Середня тривалість перебування в відповідних станах (експериментальна):")
print(tn_exp1)
print("Середні моменти першого прибуття в відповідні стани (експериментальні):")
print(t_exp1)
print("Нестабільний стан:")
print("Середня тривалість перебування в відповідних станах:")
print(tn2)
print("Середні моменти першого прибуття в відповідні стани:")
print(t2)
print("Середня тривалість перебування в відповідних станах (експериментальна):")
print(tn_exp2)
print("Середні моменти першого прибуття в відповідні стани (експериментальні):")
print(t_exp2)
```
