```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 30
lambdas = [2, 3, 4]
sam_exp_distr = [np.random.exponential(1/i, n) for i in lambdas]
sum_t = [[np.sum(X[0:i]) for i in range(n+1)] for X in sam_exp_distr]
x = np.linspace(0, n, n+1)
plt.figure(figsize=(16, 10))
plt.grid()
graphs = [plt.step(sum_t[i], x, label="Lambda = %d"%lambdas[i])[0] for i in range(len(lambdas))]
plt.legend(handles=graphs, loc=2)
plt.title("Poisson Process", fontdict={'fontname': 'Times New Roman', 'fontsize': 21}, y=1.03)
```

```python
num = 10000
sam_exp_distr_diagram = [np.random.exponential(i, num) for i in lambdas]
#k = int(1 + 3.3*math.log10(m))

for i in range(len(lambdas)):
    plt.hist(sam_exp_distr_diagram[i], edgecolor='black')
    plt.title("     lambda = {}".format(lambdas[i]) )
    plt.show()
```
