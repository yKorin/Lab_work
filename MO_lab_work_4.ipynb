```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sin(x):
     return np.sin(x**5-4*x**4+15*x**2-4)
def d_sin(x):
    return x*(5*x**3+16*x**2+30)*np.cos(x**5-4*x**4+15*x**2-4)
```

```python
x0 = (6 + 6**(1/2))/2 #x = 4.224744871391589
epsilon = 5e-09  #epsilon = 0.5 * 10**(-8)
M = 5.493562*10**(6)
h = round((3*epsilon/M)**(1/3), 10)
print("h = ", h)
#h = 0.000000001
xi = [x0 + i for i in np.linspace(-2+h, 2+h, 5)]
y = [round(sin(i), 10) for i in xi]
print("xi = ", xi,"\nyi = ", yi)
dir_y = (y[3]-y[1])/(2*h) + ((y[3]-y[1])/(2*h)-(y[4]-y[0])/(4*h))/(3)
r = M*h**2/6
print("y\' = ", round(dir_y, 8), "- наближенне значення")
print("y\' = ", round(d_sin(x0), 8), "- точне значення")
print("похибка = ", round(r, 8))
```

```python
x = np.linspace(4.2, 4.24)
plt.figure(figsize=(16, 10))
plt.grid()
plt.plot(x, sin(x), label = 'y')
plt.xlabel('x')
plt.ylabel('y')
plt.title('plot')
plt.legend(loc=2)
```

```python
x = np.linspace(4.2, 4.24)
plt.figure(figsize=(16, 20))
plt.grid()
plt.plot(x, sin(x), label = 'y')
plt.plot(x, d_sin(x), label = 'y\'')
plt.xlabel('x')
plt.ylabel('y, y\'')
plt.title('plot')
plt.legend(loc=2)
```
